Schema can be defined using StructType, StructField and types: StringType, IntegerType, FloatType, DateType etc.

```python
from pyspark.sql.types import StructType, StructField, StringType, IntegerType, BooleanType, DateType, FloatType

schema_as_string = "name string, age int, price float, orderDate date"

schema_types = StructType([
    StructField("name", StringType(), True),
    StructField("age", IntegerType(), True),
    StructField("price", FloatType(), True),
    StructField("orderDate", DateType(), True)
])
```

Having that schema we can create a DataFrame which will have following data

```python
from datetime import datetime

test_data = [
    ["Alex", 10, 33.12, datetime.strptime("2045-12-15", "%Y-%m-%d")],
    ["Agnes",77, 897.9, datetime.strptime("1999-04-12", "%Y-%m-%d")]
]
```

```python
test_df = spark.createDataFrame(data = test_data, schema = schema_as_string)
# test_df = spark.createDataFrame(data = test_data, schema = schema_types)

test_df.show()
```
