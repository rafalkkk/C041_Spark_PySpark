# Other types

| Join Type        | Alternative Name       | Description                                           |
|-----------------|----------------------|------------------------------------------------------|
| **Inner Join**   | Equi Join             | Returns only matching records from both tables.     |
| **Left Join**    | Left Outer Join       | Returns all records from the left table and matching ones from the right. Unmatched rows have NULLs. |
| **Right Join**   | Right Outer Join      | Returns all records from the right table and matching ones from the left. Unmatched rows have NULLs. |
| **Full Outer Join** | Outer Join          | Returns all records from both tables. Missing matches are filled with NULLs. |
| **Cross Join**   | Cartesian Join        | Returns the Cartesian product of both tables (every row from one table paired with every row from the other). |
| **Left Semi Join** | Semi Join            | Returns only records from the left table that have matching values in the right table but does not include columns from the right table. |
| **Left Anti Join** | Anti Join            | Returns only records from the left table that do **not** have matching values in the right table. |


```python
from pyspark.sql.types import StructType, StructField, IntegerType, StringType

students_schema = StructType([
    StructField("student_id", IntegerType(), False),
    StructField("student_name", StringType(), False)
])

students_data = [
    (1, "Alice"),
    (2, "Bob"),
    (3, "Charlie")  # Charlie has no assigned tasks
]

students_df = spark.createDataFrame(students_data, students_schema)
```

```python
tasks_schema = StructType([
    StructField("task_id", IntegerType(), False),
    StructField("task_name", StringType(), False),
    StructField("student_id", IntegerType(), True)  # Nullable to allow unassigned tasks
])

tasks_data = [
    (1, "Write an essay", 1),
    (2, "Complete math homework", 1),
    (3, "Prepare presentation", 2),
    (4, "Read assigned book", 2),
    (5, "Solve coding challenge", None),  # No student assigned
    (6, "Research project topic", None)  # No student assigned
]

tasks_df = spark.createDataFrame(tasks_data, tasks_schema)
```

```python
students_df.show()
tasks_df.show(truncate=False)
```

```python
inner_join_df = students_df.join(tasks_df, "student_id", "inner")
inner_join_df.show(truncate=False)
```

```python
left_join_df = students_df.join(tasks_df, "student_id", "left")
left_join_df.show(truncate=False)
```

```python
right_join_df = students_df.join(tasks_df, "student_id", "right")
right_join_df.show(truncate=False)
```

```python
full_join_df = students_df.join(tasks_df, "student_id", "outer")
full_join_df.show(truncate=False)
```

```python
cross_join_df = students_df.crossJoin(tasks_df)
cross_join_df.show(truncate=False)
```

```python
left_semi_join_df = students_df.join(tasks_df, "student_id", "left_semi")
left_semi_join_df.show(truncate=False)
```

```python
left_semi_join_df = students_df.join(tasks_df, "student_id", "left").filter(col("task_id").isNotNull())
left_semi_join_df.show(truncate=False)
```

```python
left_anti_join_df = students_df.join(tasks_df, "student_id", "left_anti")
left_anti_join_df.show(truncate=False)
```

```python
left_semi_join_df = students_df.join(tasks_df, "student_id", "left").filter(col("task_id").isNull())
left_semi_join_df.show(truncate=False)
```
