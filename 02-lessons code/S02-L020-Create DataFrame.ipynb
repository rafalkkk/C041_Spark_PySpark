When creating a DataFrame, we can define the schema

```python
spark.createDataFrame?
```

Internally DataFrames are implemented using Resilient Distributed Dataset (RDD)

```python
rivers.rdd.getNumPartitions()
```

Default number of partitions depends on the number of cores used by Spark

```python
conf = spark.sparkContext.getConf()
conf.get("spark.executor.memory", "unknown")
```

Data Frame transformations are lazy. They are executed first when action is called.

```python
rivers.show()
```

DataFrames are immutable

```python
rivers.where("length > 200").show()
```

```python
rivers.count()
```

```python
long_rivers = rivers.where("length > 200")
```

```python
long_rivers.show()
```

```python
long_rivers.count()
```

```python
long_rivers.rdd.getNumPartitions()
```
