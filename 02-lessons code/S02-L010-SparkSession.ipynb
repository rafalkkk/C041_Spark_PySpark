# S02-010 - Spark Session

First spark object

```python
from pyspark.sql import SparkSession
```

```python
spark = SparkSession.builder.appName("myApp").master("local[*]").getOrCreate()
```

```python
spark
```

```python
data = [ [1], [2], [3]]
df = spark.createDataFrame(data, ['id'])
```

```python
df.show()
```

Second spark object (with more parameters)

```python
spark2 = (
  SparkSession.
  builder.
  appName("myApp2").
  master("local[*]").
  config("spark.executor.memory", "2g").
  config("spark.driver.memory", "2g").
  getOrCreate()
)
```

```python
spark2
```

```python
data2 = [ [1], [2], [3]]
df2 = spark2.createDataFrame(data2, ['id'])
```

```python
df2.show()
```

Getting active spark object

```python
s = spark2.getActiveSession()
```

```python
s
```

Checking parameters

```python
spark2.conf.get("spark.executor.memory", "Unknown")
```

```python
spark.conf.get("spark.executor.memory", "Unknown")
```

```python
spark.conf.get("spark.executor.power_mode", "Unknown")
```
