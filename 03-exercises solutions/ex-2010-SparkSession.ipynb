```python
# 1. Import SparkSession
from pyspark.sql import SparkSession
```

```python
# 2. Create a new Spark session with default settings
spark = SparkSession.builder.appName("TravelDestinations").getOrCreate()
```

```python
# 3. Create a DataFrame with countries you'd like to visit
data = [("Japan", "Tokyo"), ("Italy", "Rome"), ("Australia", "Sydney")]
columns = ["Country", "Capital"]

df = spark.createDataFrame(data, columns)
```

```python
# 4. Show the table
df.show()
```

```python
# 7. Create a new session with 2GB memory per executor

from pyspark.sql import SparkSession

spark = SparkSession.builder \
    .appName("NewTravelDestinations") \
    .config("spark.executor.memory", "2g") \
    .getOrCreate()
```

```python
# 8. Recreate the DataFrame and show it again
data = [("Japan", "Tokyo"), ("Italy", "Rome"), ("Australia", "Sydney")]
columns = ["Country", "Capital"]

df = spark.createDataFrame(data, columns)

df.show()
```

```python
print(f"Application Name: {spark.sparkContext.appName}")
print(f"Master URL: {spark.sparkContext.master}")
print(f"Spark Version: {spark.version}")
print(f"Number of Executors: {spark.sparkContext.defaultParallelism}")
print(f"User Home Directory: {spark.sparkContext.sparkUser()}")
```
