```python
# Write CSV file
json_content = """{
  "colors": [
    {
      "color": "black",
      "category": "hue",
      "type": "primary",
      "code": {
        "rgba": [255,255,255,1],
        "hex": "#000"
      }
    },
    {
      "color": "white",
      "category": "value",
      "code": {
        "rgba": [0,0,0,1],
        "hex": "#FFF"
      }
    },
    {
      "color": "red",
      "category": "hue",
      "type": "primary",
      "code": {
        "rgba": [255,0,0,1],
        "hex": "#FF0"
      }
    },
    {
      "color": "blue",
      "category": "hue",
      "type": "primary",
      "code": {
        "rgba": [0,0,255,1],
        "hex": "#00F"
      }
    },
    {
      "color": "yellow",
      "category": "hue",
      "type": "primary",
      "code": {
        "rgba": [255,255,0,1],
        "hex": "#FF0"
      }
    },
    {
      "color": "green",
      "category": "hue",
      "type": "secondary",
      "code": {
        "rgba": [0,255,0,1],
        "hex": "#0F0"
      }
    }
  ]
}"""

with open("colors.json", "w") as f:
    f.write(json_content)
```

```python
import json

# Load JSON file
with open("colors.json", "r") as file:
    data = json.load(file)

# Extract relevant list
colors_data = data["colors"]

# Print sample output
print(colors_data)
```

```python
colors_data[0]
```

```python
white = None
# Search for "white"
for item in colors_data:
  if item["color"] == "white":
    white = item
    break # Exit the loop after the first match
else:
  print("Color 'white' not found.")

white
```
