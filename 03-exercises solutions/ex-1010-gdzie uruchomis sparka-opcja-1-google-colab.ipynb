```python
from pyspark.sql import SparkSession

# Sample weather data (replace with your actual observations)
data = [
    ("Poniedziałek", 15, "Pochmurno", 2),
    ("Wtorek", 18, "Słonecznie", 0),
    ("Środa", 12, "Pochmurno", 5),
    ("Czwartek", 20, "Lekkie zachmurzenie", 0),
    ("Piątek", 16, "Pochmurno", 3)
]

spark = SparkSession.builder.appName("WeatherData").getOrCreate()

# Create a DataFrame
weather_df = spark.createDataFrame(data, ["Dzień tygodnia", "Temperatura", "Zachmurzenie", "Opady (mm)"])


# Sort by temperature in descending order
df_sorted_desc = weather_df.sort(weather_df["Temperatura"].desc())
print("Dane posortowane malejąco wg temperatury:")
df_sorted_desc.show()


# Sort by temperature in ascending order
df_sorted_asc = weather_df.sort(weather_df["Temperatura"].asc())
print("\nDane posortowane rosnąco wg temperatury:")
df_sorted_asc.show()
```
